```python
def get_custom_season(row):
    month = row['month']
    day = row['day']
    
    # Define the date ranges for each season
    if (month == 2 and day >= 15) or (month == 3 and day <= 15) :
        return 'Spring'
    elif (month == 3 and day >= 16) or (month == 5) or (month == 4) or (month == 6 and day <= 7):
        return 'Summer'
    elif (month == 6 and day >= 8) or (month in [7, 8, 9]) :
        return 'Monsoon'
    elif (month == 10 ) or (month == 11 and day <= 15):
        return 'Autumn'
    else:
        return 'Winter'

# Apply the function to assign seasons
weather['custom_season'] = weather.apply(get_custom_season, axis=1)

# Calculate the average temperature by custom season
season_avg_temp = weather.groupby('custom_season')['tempC'].mean()

# Plot the average temperature by custom season
season_avg_temp.plot(kind='bar', figsize=(10, 5), title='Average Temperature by Custom Season')
plt.ylabel('Average Temperature (Â°C)')
plt.show()
```

```python


sns.boxplot(x='custom_season', y='tempC', data=weather)
plt.title('Temperature Distribution by Season')
plt.xlabel('Season')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

```python
# Loop through each season
for season in weather['custom_season'].unique():
    print(f"Correlation matrix for {season}:")
    
    # Filter the data for the current season
    season_data = weather[weather['custom_season'] == season]
    
    # Select only numeric columns (int64 and float64 types)
    numeric_data = season_data.select_dtypes(include=['int64', 'float64'])
    
    # Calculate the correlation matrix
    correlation_matrix = numeric_data.corr()
    
    # Print the correlation matrix
    print(correlation_matrix)
    print("\n")
```
